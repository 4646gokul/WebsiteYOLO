```python
train_gt_path = './data/wider_face_split/wider_face_train_bbx_gt.txt'
# train_gt_path = './data/wider_face_split/readme.txt'

meta = {}

with open(train_gt_path, 'r') as rfile:
    
    i = 0
    while True:
        short_file_path = rfile.readline(); i += 1
        bbox_count = rfile.readline(); i += 1
        
        short_file_path = short_file_path.replace('\n', '')
        
        try:
            bbox_count = int(bbox_count.replace('\n', ''))
            
        except ValueError:
            print(i, short_file_path, bbox_count)
        
        gt_bboxes = []
        
        for _ in range(bbox_count):
            attributes = rfile.readline(); i += 1
            attributes = attributes.replace('\n', '').split(' ')
            attributes = [int(att) for att in attributes if len(att) > 0]

            gt_bboxes.append(attributes)

        meta[short_file_path] = gt_bboxes
```
